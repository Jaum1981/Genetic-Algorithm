# Definição da Classe Problem

**__init__(self, adjacency_matrix)**

* Parâmetros:

  * adjacency_matrix: Uma matriz de adjacência representando o grafo. Cada elemento (i, j) é 1 se existe uma aresta entre os vértices i e j, e 0 caso contrário.

* Atributos:

  * self.adjacency_matrix: Armazena a matriz de adjacência.
  * self.num_vertices: Número de vértices no grafo.

* Inicialização do Problema:

  * n_var=self.num_vertices: O número de variáveis é igual ao número de vértices. Cada variável representará a inclusão ou exclusão de um vértice na clique.
  * n_obj=1: Há um único objetivo, que é maximizar o tamanho da clique.
  * xl=0 e xu=1: Limites inferior e superior para as variáveis, indicando que elas podem assumir valores contínuos entre 0 e 1.
  * type_var=float: As variáveis são do tipo contínuo (float).

**_evaluate(self, X, out, *args, **kwargs)**

* Parâmetros:

  * X: Conjunto de soluções a serem avaliadas. Cada solução é um vetor contínuo representando uma possível configuração da clique.
out: Dicionário onde os resultados da avaliação serão armazenados.

* Processo:

  Avalia cada solução em X usando a função _fitness e armazena o resultado no dicionário out sob a chave "F".


**_fitness(self, x)**

* Parâmetros:

  * x: Um vetor contínuo representando uma solução candidata.

* Processo:

  Converte o vetor contínuo em uma clique usando a função decode.
  Calcula o tamanho da clique e retorna o valor negativo desse tamanho. O valor negativo é utilizado porque a função de fitness é geralmente minimizada, então maximizar o tamanho da clique é equivalente a minimizar seu valor negativo.


**decode(self, x)**

* Parâmetros:

  * x: Vetor contínuo representando uma solução.

* Processo:

  Ordena os vértices com base nos valores no vetor contínuo. O uso de np.argsort(-x) garante que os vértices com maiores valores contínuos vêm primeiro.
  Tenta construir uma clique adicionando vértices na ordem obtida. A função _is_clique_valid é usada para verificar se a clique é válida antes de adicionar um novo vértice.

* Retorno:

  * Retorna a lista de vértices que formam a clique.


```python
import numpy as np
from pymoo.algorithms.soo.nonconvex.brkga import BRKGA
from pymoo.optimize import minimize
from pymoo.core.problem import Problem
from pymoo.termination import get_termination

import numpy as np
from pymoo.core.problem import Problem

class MaxCliqueProblem(Problem):

    # matriz de adjacência vai ser o grafo
    def __init__(self, adjacency_matrix):
        self.adjacency_matrix = adjacency_matrix
        self.num_vertices = adjacency_matrix.shape[0]

        # max sum x[i]
        #x[i] = 1, se o vertice i está na clique 0, caso contrario

        # Inicializar o problema com uma dimensão igual ao número de vértices
        super().__init__(n_var=self.num_vertices, n_obj=1, xl=0, xu=1, type_var=float)

    def _evaluate(self, X, out, *args, **kwargs):
        # Avaliação das soluções em X
        out["F"] = np.array([self._fitness(x) for x in X])

    def _fitness(self, x):
        # Retorna o valor de fitness (tamanho da clique) para uma solução x
        clique = self.decode(x)  # Usar a função decode para obter a clique
        return -len(clique)  # Fitness é o tamanho da clique (negativo para minimizar)

    def decode(self, x):
        """
        Transforma um vetor contínuo em um conjunto de vértices que forma uma clique.

        Parameters:
        - x (np.ndarray): Vetor contínuo representando a solução.

        Returns:
        - clique (list): Lista de vértices que formam a clique.
        """
        # Obter a ordem dos vértices com base nos valores contínuos
        ordered_vertices = np.argsort(-x)  # Ordena em ordem decrescente

        clique = []

        # Construir a clique a partir da ordem dos vértices
        for vertex in ordered_vertices:
            if self._is_clique_valid(clique + [vertex]):
                clique.append(vertex)

        return clique

    def _is_clique_valid(self, clique):
        """
        Verifica se o conjunto de vértices forma uma clique válida.

        Parameters:
        - clique (list): Lista de vértices para verificar se formam uma clique.

        Returns:
        - bool: True se formar uma clique, False caso contrário.
        """
        clique_size = len(clique)
        for i in range(clique_size):
            for j in range(i + 1, clique_size):
                if self.adjacency_matrix[clique[i], clique[j]] == 0:
                    return False
        return True
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import time

# Parâmetros do grafo
num_vertices = 80
density = 0.9

# Criação do grafo aleatório
G = nx.erdos_renyi_graph(num_vertices, density)

# Exibindo o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Layout para o posicionamento dos nós
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=16, font_weight='bold', edge_color='gray')
plt.title(f'Grafo Aleatório com {num_vertices} Vértices e Densidade {density}')
plt.show()

# Obtendo e mostrando a matriz de adjacência
adjacency_matrix = nx.adjacency_matrix(G).todense()  # Obtém a matriz de adjacência
print("Matriz de Adjacência:")
print(np.array(adjacency_matrix))

# Calculando a clique máxima
start_time = time.time()
max_clique = nx.algorithms.clique.find_cliques(G)
# Encontrar a maior clique
max_clique = max(max_clique, key=len)
end_time = time.time()

# Exibindo o resultado da clique máxima
print(f"\nMaior Clique Encontrada: {max_clique}")
print(f"Tamanho da Maior Clique: {len(max_clique)}")
print(f"Tempo de Cálculo: {end_time - start_time:.4f} segundos")
```

# Parâmetros do BRKGA

* n_elites

  * Descrição: O número de indivíduos que são selecionados como elites na população. Esses indivíduos são garantidos para serem preservados entre as gerações.
  * Tipo: Inteiro
  * Uso: Controla a quantidade de soluções de alta qualidade que são preservadas para garantir que as melhores soluções encontradas não sejam perdidas.

* n_offsprings

  * Descrição: O número de indivíduos (ou filhos) gerados por crossover a partir dos pais na população.
  * Tipo: Inteiro
  * Uso: Define o tamanho da nova população gerada por operadores de crossover. Mais offsprings podem aumentar a diversidade na população.

* n_mutants

  * Descrição: O número de indivíduos na população que são gerados aleatoriamente por mutação.
  * Tipo: Inteiro
  * Uso: Introduz diversidade na população para explorar novas áreas do espaço de busca. O número de mutants pode ajudar a evitar a convergência prematura.

* p_mut

  * Descrição: A probabilidade de mutação aplicada a um indivíduo.
  * Tipo: Float (entre 0 e 1)
  * Uso: Controla a intensidade das mutações. Um valor maior aumenta a chance de mutações, promovendo maior exploração.

* bias

  * Descrição: O parâmetro de viés para o algoritmo BRKGA, que controla a influência do componente aleatório versus o componente baseado em elitismo.
  * Tipo: Float (entre 0 e 1)
  * Uso: Ajusta a probabilidade de um indivíduo ser selecionado com base em sua qualidade relativa, ajudando a balancear a exploração e a exploração.

* eliminate_duplicates

  * Descrição: Booleano que indica se soluções duplicadas devem ser eliminadas na população.
  * Tipo: Booleano
  * Uso: Se True, evita a duplicação de soluções na população, o que pode melhorar a diversidade e eficiência do algoritmo.

# Parâmetros Opcionais

* sampling

  * Descrição: Um método de amostragem personalizado para gerar indivíduos na população inicial.
  * Tipo: Customizado (por exemplo, uma instância de uma classe de amostragem)

* output

  * Descrição: Um método personalizado para definir o formato e o tipo de saída do algoritmo.
  * Tipo: Customizado (por exemplo, uma instância de uma classe de saída)

* crossover

  * Descrição: O operador de crossover personalizado a ser usado no algoritmo.

# CustomCrossover

A classe CustomCrossover é uma implementação personalizada do operador de crossover para o BRKGA (Biased Random Key Genetic Algorithm). O crossover é uma etapa importante em algoritmos evolutivos onde duas soluções (pais) são combinadas para gerar novas soluções (descendentes).

* Componentes da Classe
  * Construtor (__init__):
    * prob: A probabilidade de aplicar o crossover. Geralmente é definida como 1.0 (ou 100%) para garantir que o crossover será aplicado sempre que necessário.
    * n_offsprings: O número de descendentes a serem gerados a partir dos pais selecionados. Neste caso, são gerados dois descendentes por operação de crossover.
    * super().__init__(prob, n_offsprings): Chama o construtor da classe base Crossover para inicializar as propriedades básicas do crossover.

  * Parâmetros:

    * problem: A instância do problema, geralmente não usada diretamente neste método, mas pode ser útil para acessar informações do problema se necessário.
    * pop: A matriz da população, onde cada linha é um vetor de solução (indivíduo).
  * parents: Os índices dos pais selecionados para o crossover.

  * Processo de Crossover:

    * offsprings: Matriz para armazenar as novas soluções geradas pelo crossover. Inicialmente, é uma matriz de zeros com o número de descendentes e o tamanho dos vetores de solução.

  * Loop de Crossover:
    * for i in range(self.n_offsprings): Itera sobre o número de descendentes a serem gerados.
    * p1, p2 = parents[i*2], parents[i*2 + 1]: Seleciona dois pais para o crossover.
    * crossover_point = np.random.randint(1, pop.shape[1]): Define um ponto de crossover aleatório. Este ponto determina onde a combinação dos pais ocorre.
    * offsprings[i, :crossover_point] = pop[p1, :crossover_point]: Copia a parte do vetor de solução do primeiro pai até o ponto de crossover.
    * offsprings[i, crossover_point:] = pop[p2, crossover_point:]: Copia a parte do vetor de solução do segundo pai a partir do ponto de crossover.

  * Retorno:

    * offsprings: A matriz com as novas soluções geradas pelo crossover.


```python
import numpy as np
from pymoo.core.crossover import Crossover

class CustomCrossover(Crossover):
    def __init__(self, prob=1.0, n_offsprings=2):
        super().__init__(prob, n_offsprings)

    def _do(self, problem, pop, parents, **kwargs):
        """
        Perform crossover on the selected parents to produce offsprings.

        Parameters:
        - problem: The problem instance.
        - pop: The population object.
        - parents: The indices of the parents selected for crossover.

        Returns:
        - offsprings: The newly created offspring solutions.
        """
        offsprings = np.zeros((self.n_offsprings, pop.shape[1]), dtype=float)

        for i in range(self.n_offsprings):
            p1, p2 = parents[i*2], parents[i*2 + 1]
            crossover_point = np.random.randint(1, pop.shape[1])
            offsprings[i, :crossover_point] = pop[p1, :crossover_point]
            offsprings[i, crossover_point:] = pop[p2, crossover_point:]

        return offsprings
```
